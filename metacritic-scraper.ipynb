```python
liste = []
```

```python
chrome_part = "C:/USERS/MERT/DESKTOP/chromedriver"
driver = webdriver.Chrome(chrome_part)
```

```python

driver.get("https://www.metacritic.com/browse/games/score/metascore/all/all/filtered?page=10")
```

```python
#1-5 arası
follow_loop = range(1,10,2)
for x in follow_loop:
    xpath = '//*[@id="main_content"]/div[1]/div[2]/div/div[1]/div/div[2]/table/tbody/tr['
    xpath += str(x)
    xpath += ']/td[2]/a/h3'
    try:
        driver.find_element_by_xpath(xpath).click()
    except:
        None
    try:
        platform = driver.find_element_by_xpath(".//*[@class='platform']").text
    except:
        None
    try:
        name = driver.find_element_by_xpath(".//*[@class='hover_none']").text
    except:
        None
    try:
        date = driver.find_element_by_xpath('//*[@id="main"]/div/div[1]/div[1]/div[1]/div[3]/ul/li[2]').text
    except:
        None
    try:
        genre = driver.find_element_by_xpath(".//*[@class='summary_detail product_genre']").text
    except:
        None
    try:    
        rating = driver.find_element_by_xpath(".//*[@class='summary_detail product_rating']").text
    except:
        None
    try:
        multi = driver.find_element_by_xpath(".//*[@class='summary_detail product_players']").text
    except:
        None
    
    try:
        meta_score = driver.find_element_by_xpath(".//*[@itemprop='ratingValue']").text
    except:
        None
    try:
        user_score = driver.find_element_by_xpath(".//*[@class='score_summary']").text
    except:
        None

    
    time.sleep(2)
    driver.execute_script("window.history.go(-1)")
    liste.append([name, date,platform,genre,multi,meta_score,user_score,rating])
```
